There is no "event_data" key in this dataset. The h5 files with "event_data" are a very old format that we no longer use. It's not possible for us to use that old format with newer datasets because it results in extremely large files. The event_data were the images as they are fed into the CNN, but the new format stores the data in a much more efficient way and then the data loader creates the images from that.

These hit files are used in the reconstruction of the image to feed the CNN - the image is generated dynamically. This is more efficient than the previous method of storing image data as "event_data" matrices.
